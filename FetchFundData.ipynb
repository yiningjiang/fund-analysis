```python
name5=data5.iloc[:,1].values
url5=data5.iloc[:,2].values
name10=data10.iloc[:,1].values
url10=data10.iloc[:,2].values
name20=data20.iloc[:,1].values
url20=data20.iloc[:,2].values
name100=data100.iloc[:,1].values
url100=data100.iloc[:,2].values
namewq=datawq.iloc[:,1].values
urlwq=datawq.iloc[:,2].values
```

```python
def get_content(driver,url):
    driver.get(url)
    time.sleep(10)
    content = driver.page_source.encode('utf-8')
    soup = BeautifulSoup(content,'lxml')
    return soup
```

```python
#抓取cta风格基金数据
for i in np.arange(len(name5)):
    table = pic1[i].findAll("table")[1]
    rows = table.findAll("tr")
    csvFile = open("data/cta/%s.csv"%name5[i], 'wt')
    writer = csv.writer(csvFile)
    try:
        tit=['净值时间','单位净值','复权净值','增长率']
        writer.writerow(tit)
        for row in rows:
            csvRow = []
            csvRow.append(re.findall(r"\d+\-\d+\-\d+",row.findAll('td')[0].get_text())[0])
            csvRow.append(re.findall(r"\d+\.\d+",row.findAll('td')[1].get_text())[0])
            csvRow.append(re.findall(r"\d+\.\d+",row.findAll('td')[2].get_text())[0])
            csvRow.append(re.findall(r"\d+\.\d+\%",row.findAll('td')[3].get_text())[0])
            writer.writerow(csvRow)
    finally:
        csvFile.close()
```
